```python
# Question 6: Write a function to demonstrate the behavior of mutable and immutable arguments

Samplelist = [1, 2, 3]
Sampletuple = (1, 2, 3)

print("Original:")
print("List:", Samplelist)
print("Tuple:", Sampletuple)

UpdateValue = int(input("Insert Updated Value: "))
UpdatePosition = int(input("Insert Updated Position: "))

def mutable_immutable_demo(Samplelist, Sampletuple):
  print("Updated:")
  Samplelist[UpdatePosition] = UpdateValue
  print("List:", Samplelist)
  try:
    Sampletuple[UpdatePosition] = UpdateValue
  except TypeError as e:
    print("Error modifying tuple:", e)

mutable_immutable_demo(Samplelist, Sampletuple)
```

```python
# Question 9: Write a Python program to convert user input from string to integer, float, and boolean types

user_input = str(input("Enter a value: "))

def convert_input():
  try:
    integer_value = int(user_input)
    print("Integer:", integer_value)
  except ValueError:
    print("Could not convert to integer.")

  try:
    float_value = float(user_input)
    print("Float:", float_value)
  except ValueError:
    print("Could not convert to float.")

  try:
    boolean_value = bool(user_input)
    print("Boolean:", boolean_value)
  except ValueError:
    print("Could not convert to boolean.")

convert_input()
```

```python
# Question 10: Write code to demonstrate type casting with list elements

Samplelist = [1.5, '2', 3, '4.5']

# Type casting list elements to integers
int_list = [int(float(x)) for x in Samplelist]
print("Integer List:", int_list)

# Type casting list elements to floats
float_list = [float(x) for x in Samplelist]
print("Float List:", float_list)

# Type casting list elements to strings
str_list = [str(x) for x in Samplelist]
print("String List:", str_list)
```

```python
# Question 13: Write a Python program to find the sum of all even numbers between 1 and 50

sum_of_evens = 0

for number in range(1, 51):
  if number % 2 == 0:
    sum_of_evens += number

print("The sum of all even numbers between 1 and 50 is:", sum_of_evens)
```

```python
# Question 14:  Write a program to reverse a string using a while loop0

def reverse_string(s):
    reversed_str = ""
    index = len(s) - 1
    while index >= 0:
        reversed_str += s[index]
        index -= 1
    return reversed_str

input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print(f"Reversed string: {reversed_string}")
```

```python
# question 15: Write a Python program to calculate the factorial of a number provided by the user using a while loop

def factorial(n):
    result = 1
    while n > 0:
        result *= n
        n -= 1
    return result

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    fact = factorial(num)
    print(f"The factorial of {num} is {fact}.")
```
